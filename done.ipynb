```python
import numpy as np
import pandas as pd

df = pd.read_csv('usage_report_full_cleaned.csv')
df.head()
```

```python
df['scheduler'].unique()
```

```python
cols = ['resource_request_cpus',
    'resource_request_memory',
    'average_usage_cpus',
    'average_usage_memory',
    'maximum_usage_cpus',
    'maximum_usage_memory']
```

```python
for col in cols:
    Q1 = df[col].quantile(0.25)
    Q2 = df[col].quantile(0.75)
    IQR = Q2 - Q1

    df[col + '_outlier_flag'] = ((df[col] < Q1 - 1.5*IQR ) | (df[col] < Q2 + 1.5*IQR)).astype(int)
```

```python
df.head()
```

```python
df[cols].skew()
```

```python
for col in cols:
    df[col] = np.log1p(df[col])
```

```python
for col in cols:
    lower = df[col].quantile(0.01)
    upper = df[col].quantile(0.99)
    df[col] = df[col].clip(lower, upper)
```

```python
df.head()
```

```python
df[cols].skew()
```

```python
df.head()
```

```python
y = df['failed']
```

```python
num_cols = [
    'resource_request_cpus',
    'resource_request_memory'
]

flag_cols = [col for col in df.columns if col.endswith('_outlier_flag')]

cat_cols = [
    'scheduler',
    'priority',
    'scheduling_class'
]
```

```python
from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()

for col in cat_cols:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])
```

```python
X = df_encoded[num_cols + flag_cols + cat_cols]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 42, stratify = y)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
dt = DecisionTreeClassifier(random_state = 42)
dt.fit(X_train,y_train)
y_pred = dt.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
y_train_pred = dt.predict(X_train)
print(classification_report(y_train, y_train_pred))
```

```python
dt.get_depth(), dt.get_n_leaves()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
dt = DecisionTreeClassifier(max_depth=14,
    min_samples_leaf=100,
    min_samples_split=100,
    random_state=42)
dt.fit(X_train,y_train)
y_pred = dt.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
dt = DecisionTreeClassifier(max_depth=14,
    min_samples_leaf=100,
    min_samples_split=200,
    random_state=42)
dt.fit(X_train,y_train)
y_pred = dt.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
dt = DecisionTreeClassifier(max_depth=20,
    min_samples_leaf=100,
    min_samples_split=200,
    random_state=42)
dt.fit(X_train,y_train)
y_pred = dt.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_samples = 0.8, oob_score = True, random_state = 42, class_weight = 'balanced')
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [50,80],
    'max_samples' : [0.8],
    'max_features': [5,10],
    'max_depth' : [15]
}
gscv = GridSearchCV(RandomForestClassifier(oob_score = True, random_state = 42),
                    param_grid = param_grid,
                    verbose = 2,
                    cv = 2)

gscv.fit(X_train,y_train)
```

```python
y_pred = gscv.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
def accuracy_fun(obj):
    obj.fit(X_train,y_train)
    y_pred = obj.predict(X_test)
    # print("Max depth: ",depth)
    from sklearn.metrics import classification_report
    print(classification_report(y_test,y_pred))
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('borg_traces_data.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df = df[['scheduling_class','priority','scheduler','collection_type','vertical_scaling',
         'start_after_collection_ids','resource_request','assigned_memory','constraint',
         'average_usage','maximum_usage','page_cache_memory','cpu_usage_distribution',
         'tail_cpu_usage_distribution','sample_rate','cycles_per_instruction','memory_accesses_per_instruction']]
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
mean_cycle=df['cycles_per_instruction'].mean()
mean_access = df['memory_accesses_per_instruction'].mean()
mode_scheduler = df['scheduler'].mode()
mode_verticalScaling = df['vertical_scaling'].mode()
```

```python
print(mode_scheduler)
```

```python
df['cycles_per_instruction'].fillna(mean_cycle,inplace=True)
df['memory_accesses_per_instruction'].fillna(mean_access,inplace=True)
df['scheduler'].fillna(mode_scheduler,inplace=True)
df['vertical_scaling'].fillna(mode_verticalScaling,inplace=True)
```

```python
df.dropna(subset=['scheduler', 'vertical_scaling', 'resource_request'], inplace=True)
```
